```python
np.random.seed(80)

rand_array = np.random.randint(0, 101, (4, 4))
print(rand_array)


print('max', rand_array.max())  # max
print('min', rand_array.min())  # min
print('argmax', rand_array.argmax())
print('argmin', rand_array.argmin())
print('shape', rand_array.shape)  # shape of the array
```

```python

print(rand_array[rand_array > 50])
```

```python
sum_columns = np.sum(rand_array, axis=0)
sum_rows = np.sum(rand_array, axis=1)
print(sum_rows)
print(sum_columns)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from PIL import Image
```

```python
image_url = "https://raw.githubusercontent.com/pythonai250824/05.03.2025/main/jinny.jpg"
image = io.imread(image_url)
```

```python
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
print("Image shape:", image.shape)
```

```python
first_50_columns = image[:, :50] 
plt.imshow(first_50_columns)
plt.title("First 50 Columns of Image")
plt.axis("off")
plt.show()
```

```python
image_pil = Image.fromarray(image) 
rotated_image = image_pil.rotate(120)
```

```python
plt.imshow(rotated_image)
plt.title("Rotated Image (120 degrees)")
plt.axis("off")
plt.show()
```
